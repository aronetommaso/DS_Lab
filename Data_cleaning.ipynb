```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Definiamo quali sono i valori corretti
valori_validi = [1, 2, 3, 4, 5]

# 2. Contiamo tutto ciò che NON è in quella lista (inclusi i NaN)
# La tilde (~) inverte la condizione. .sum() conta i True per ogni colonna.
missing_counts = (~att_beh.isin(valori_validi)).sum()

# (Opzionale) Se vuoi visualizzare solo le colonne che hanno effettivamente valori mancanti:
# missing_counts = missing_counts[missing_counts > 0]

# 3. Creazione dell'Istogramma (Bar Chart)
plt.figure(figsize=(16, 7)) # Dimensioni larghe per far entrare bene tutte e 37 le colonne

# Creiamo le barre con un bel colore di allerta per i missing values
ax = missing_counts.plot(kind='bar', color='crimson', edgecolor='black', alpha=0.8)

# 4. Personalizzazione estetica per la massima chiarezza
plt.title("Frequenza dei Missing Values (-99, -98, -97, NaN) per Variabile", fontsize=16, fontweight='bold', pad=20)
plt.xlabel("Variabili (Domande del Sondaggio)", fontsize=12, fontweight='bold')
plt.ylabel("Numero di valori anomali/mancanti", fontsize=12, fontweight='bold')

# Ruotiamo le etichette dell'asse X per renderle leggibili
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)

# Aggiungiamo una griglia orizzontale di riferimento
plt.grid(axis='y', linestyle='--', alpha=0.5)

# 5. Il tocco finale: scriviamo il numero esatto in cima a ogni barra
for p in ax.patches:
    altezza = p.get_height()
    if altezza > 0: # Scrive il numero solo se c'è almeno un valore mancante
        ax.annotate(f'{altezza}', 
                    (p.get_x() + p.get_width() / 2., altezza), 
                    ha='center', va='bottom', 
                    xytext=(0, 5), 
                    textcoords='offset points',
                    fontsize=9, fontweight='bold', color='darkred')

# Ottimizza gli spazi per non tagliare le etichette in basso
plt.tight_layout()

# Mostra il grafico
plt.show()

# Stampa anche i numeri grezzi come testo sotto al grafico
print("--- Riepilogo Testuale dei Valori Mancanti ---")
print(missing_counts.to_string())
```

```python
import pandas as pd
import numpy as np

df_calc = att_beh.copy()
df_calc = df_calc.where(df_calc.isin([1, 2, 3, 4, 5]))  #considero solo le risposet presenti non NA

# per le varaibili rosse, (una risposta esatta è quando il punteggio è basso) applico la trasformazione y=6-x 
#quindi se uno ha risposto 1 è come se avesse risposto 5 (la perfezione) viceversa se avesse rispsto 5 diventerebbe 1 (risposta pessima)
for col in red_attributes:
    if col in df_calc.columns:
        df_calc[col] = 6 - df_calc[col]

#per le variabili neutre non le cosnidero nel conteggio il loro punteggio è 0 
for col in green_attributes:
    if col in df_calc.columns:
        df_calc[col] = 0

#Calcolo degli Score Finali
financial_score_df = pd.DataFrame({
    'i_esimo': df_calc.index,
    'finacial_situation': df_calc[blue_cat].sum(axis=1),
    'behaviour_investement-payment': df_calc[pink_cat].sum(axis=1),
    'knowledge_financial_privacy_digital': df_calc[brown_cat].sum(axis=1)
})

# Calcolo la percentuale: (numero risposte valide / numero totale domande della categoria) * 100

financial_score_df['perc_risposte_situation'] = (df_calc[blue_cat].count(axis=1) / len(blue_cat)) * 100
financial_score_df['perc_risposte_behaviour'] = (df_calc[pink_cat].count(axis=1) / len(pink_cat)) * 100
financial_score_df['perc_risposte_knowledge'] = (df_calc[brown_cat].count(axis=1) / len(brown_cat)) * 100

# Stampa per controllare le nuove colonne
financial_score_df.head()

#financial_score_df
#print(financial_score_df.head())
```

```python
import pandas as pd

def calcola_e_sostituisci_score(df):
      
    qs_cols = [col for col in df.columns if str(col).startswith('qs')] #tutte le variabili Attitudes and Behaviour iniziano con qs
    
    # Categoria (Polarità e Categorie)
    yellow_attributes=['qs1_1','qs1_3', 'qs1_7', 'qs1_10', 'qs1_13', 'qs2_1', 'qs2_2', 'qs2_6', 'qs2_8','qs3_3','qs3_9','qs3_10', 'qs3_11', 'qs3_12', 
                   'qs4_1', 'qs4_3', 'qs4_7', 'qs4_8', 'qs5_4', 'qs5_5', 'qs5_6']
    red_attributes = ['qs1_2', 'qs1_5', 'qs1_8', 'qs1_9', 'qs2_3', 'qs2_4', 'qs2_5', 'qs2_7', 'qs3_13', 'qs4_2', 'qs4_4', 'qs4_5', 'qs4_6']
    green_attributes = ['qs3_2', 'qs2_9', 'qs1_4']
    
    pink_cat = ['qs1_1', 'qs1_2', 'qs1_3', 'qs1_5', 'qs1_8', 'qs1_13', 'qs2_1', 'qs2_3', 'qs2_5', 'qs2_9', 'qs3_2', 'qs3_11', 'qs3_12', 'qs4_7', 'qs5_4', 'qs5_5', 'qs5_6']
    brown_cat = ['qs1_9', 'qs2_6', 'qs2_7', 'qs2_8', 'qs3_13', 'qs4_1', 'qs4_2', 'qs4_3', 'qs4_4', 'qs4_5', 'qs4_6', 'qs4_8']
    blue_cat = ['qs1_4', 'qs1_7', 'qs1_10', 'qs2_2', 'qs2_4', 'qs3_3', 'qs3_9', 'qs3_10']

    # considero solo le variabili d'interesse e NON NAN
    df_calc = df[qs_cols].copy()
    df_calc = df_calc.where(df_calc.isin([1, 2, 3, 4, 5]))
    
    # trasformazione variabili rosse e verdi
    for col in red_attributes:
        if col in df_calc.columns:
            df_calc[col] = 6 - df_calc[col]
            
    for col in green_attributes:
        if col in df_calc.columns:
            df_calc[col] = 0
            
    # creo il nuovo dataset senza le variabili  'qs'
    df_finale = df.drop(columns=qs_cols).copy()
    
    # aggiungo le tre variabili nuove
    df_finale['finacial_situation'] = df_calc[blue_cat].sum(axis=1)
    df_finale['behaviour_investement-payment'] = df_calc[pink_cat].sum(axis=1)
    df_finale['knowledge_financial_privacy_digital'] = df_calc[brown_cat].sum(axis=1)
    

    return df_finale
```
