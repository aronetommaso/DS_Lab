```python
import pandas as pd
import numpy as np

# 1. Caricamento del dataset originale
df = pd.read_csv('Database_ENG.csv')

# 2. Definizione dei codici tecnici di "Missing" secondo il manuale IACOFI 2023
technical_na_codes = [-95,-99, -98, -97, -999]

# 3. Generazione di df_cleaned
df_cleaned = df.replace(technical_na_codes, np.nan)

# 4. Conteggio NA per COLONNA (Variabili/Domande)
na_per_colonna = df_cleaned.isna().sum()

# 5. Conteggio NA per RIGA (Singoli Intervistati)
na_per_riga = df_cleaned.isna().sum(axis=1)
nariga = (na_per_riga/len(df_cleaned.columns)).sort_values(ascending=False)
nacol = (na_per_colonna/len(df_cleaned)).sort_values(ascending=False)
```

```python
# Calcola la percentuale di valori NaN sull'intero dataset
percentuale_na_cleaned = df_cleaned.isna().sum().sum() / (df.shape[0] * df.shape[1]) * 100
print(f"Percentuale di NA nel dataset: {percentuale_na_cleaned:.2f}%")
```

```python
# Calcola la percentuale di valori NaN sull'intero dataset
percentuale_na = df.isna().sum().sum() / (df.shape[0] * df.shape[1]) * 100
print(f"Percentuale di NA nel dataset: {percentuale_na:.2f}%")
```
